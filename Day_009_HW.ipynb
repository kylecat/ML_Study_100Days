# 檢視與處理 Outliers
### 為何會有 outliers, 常見的 outlier 原因
* 未知值，隨意填補 (約定俗成的代入)，如年齡常見 0,999
* 可能的錯誤紀錄/手誤/系統性錯誤，如某本書在某筆訂單的銷售量 = 1000 本

```python
# Import 需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import lag_plot

# 讓圖片不會另開視窗彈出來
%matplotlib inline

# 設定 data_path
dir_data = './data'
```

```python
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
app_train.head()  # 先看一下資料
```

## 請參考 HomeCredit_columns_description.csv 的欄位說明，觀察並列出三個你覺得可能有 outlier 的欄位並解釋可能的原因

```python
### 先篩選數值型的欄位
"""
YOUR CODE HERE, fill correct data types (for example str, float, int, ...)
"""
# print(app_train.dtypes)

dtype_select = ['float64','int64']

colums_dtypes = app_train.dtypes    # 找出適合符合dtypes格式的columns

# app_train.dtypes.isin(dtype_select) 有bug 只能找到'int64'，所以另外找方法做確認用的list
list_dtypes = [True if c in dtype_select else False for c in colums_dtypes]
# print(list_dtypes) # Debug

# 篩選後的比對list，再跟原本的資料比對後提出能用的欄位
numeric_columns = app_train.columns[list_dtypes].tolist()  
# print(numeric_columns) # Debug

# 再把只有 2種數值 (通常是 0,1) 的欄位去掉
# print(app_train[numeric_columns].apply(lambda x:len(x.unique())>2 )) # Debug
numeric_columns = list(app_train[numeric_columns].columns[list(app_train[numeric_columns].apply(lambda x:len(x.unique())>2 ))])
print(numeric_columns)
print("Numbers of remain columns {}".format(len(numeric_columns)))
```

```python
# 檢視這些欄位的數值範圍
# 用value_counts()來看一下個數職的重複情況及分布
# 如果有離群子/異常值，則出現的型態會跟其他數值不一樣

for col in numeric_columns:
    """
    Your CODE HERE, make the box plot
    """
#     print(col)
    v_counts=app_train[col].value_counts()
    plt.title(col)
    plt.xlabel("Value")   # 欄位裡獨立的數值
    plt.ylabel("Counts")  # 出現次數
    plt.scatter(v_counts.index, v_counts)
#     lag_plot(data)
    plt.show()
```

### 從上面的圖檢查的結果，至少這三個欄位好像有點可疑

* AMT_INCOME_TOTAL
* REGION_POPULATION_RELATIVE
* OBS_60_CNT_SOCIAL_CIRCLE

### Hints: Emprical Cumulative Density Plot, [ECDF](https://zh.wikipedia.org/wiki/%E7%BB%8F%E9%AA%8C%E5%88%86%E5%B8%83%E5%87%BD%E6%95%B0), [ECDF with Python](https://stackoverflow.com/questions/14006520/ecdf-in-python-without-step-function)

```python
# 最大值離平均與中位數很遠
print(app_train['AMT_INCOME_TOTAL'].describe())

# 繪製 Empirical Cumulative Density Plot (ECDF)
"""
YOUR CODE HERE
"""
def DEF(_Series):
    # Series.value_counts() 先讓Series統計一下個數值出現的次數(按照數值出現順序排列)
    print(type(_Series))
    _data = _Series.value_counts()
    
    # 因為data是用value作排列，所以再用一次 Series.sort_index() 讓data 用index來排序
    # 這時候可以得到　app_train['AMT_INCOME_TOTAL']數值排序後的出現次數
    _data = _data.sort_index()
    # print(type(data))
    # print(data)
    
    # 最後用累加function　Series.cumsum()　算出cdf值 
    _data = _data.cumsum()
    
    return _data

# 最後用累加function　Series.cumsum()　算出cdf值 
cdf = DEF(app_train['AMT_INCOME_TOTAL'])
# print(cdf)
```

```python
plt.plot(list(cdf.index), cdf/cdf.max())
plt.xlabel('Value')
plt.ylabel('ECDF')
plt.xlim([cdf.index.min(), cdf.index.max() * 1.05]) # 限制顯示圖片的範圍
plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍

plt.show()
```

```python
# 畫完發現其實不是很清楚(數值上升太快根本看不出來)
# 改變 y 軸的 Scale, 讓我們可以正常檢視 ECDF
# plt.plot(x,y)

plt.plot(np.log(list(cdf.index)), cdf/cdf.max())
plt.xlabel('Value (log-scale)')
plt.ylabel('ECDF')

plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍

plt.show()
```
